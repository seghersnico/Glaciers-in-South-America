Interpretation of the combined PSD plot:
- The dark green line represents the Power Spectral Density (PSD) of the glacier mass balance data, calculated using Welch's method.
- The Y-axis shows 'Power Spectral Density', which is a measure of energy per frequency unit.
- Peaks in the PSD plot are generally 'smoother' and more reliable than those from a single FFT, thanks to segmentation and averaging.
- The blue and magenta lines are the scaled Kernel Density Estimates (KDEs) of your specified ENSO and PDO return periods.
- Where the KDE peaks overlap with the PSD peaks, there is a strong correlation between the periodicity of your climate events and the energy distribution in the glacier mass balance.
  Note: The scaling factors for the KDEs are adjusted based on the expected PSD values for better visual comparison.
  You can experiment with the '* 0.7' scaling factor in the KDE scaling helper function to adjust the relative height for optimal visual comparison.
